```python
class CustomLinearRegression:
  def __init__(self):
    self.m = None
    self.c = None

  def fit(self , x , y):
    p , q = 0 , 0
    for i in range(x.shape[0]):
      p += (y[i] - y.mean()) * (x[i] - x.mean()) 
      q += (x[i] - x.mean()) ** 2
    self.m = p/q 
    self.c = y.mean() - m * x.mean()

    return self.m , self.c
  def predict(self,x):
    return self.m * x + self.c
```

```python
# gradient descent
import numpy as np
class CustomLinearRegressionGD:

  def __init__(self , learning_rate , iter):
    self.learning_rate = learning_rate
    self.iter = iter

  def fit(self , X , Y):
    " X-> X_train , Y-> y_train"
    row , col = X.shape
    self.W = np.random.rand( col )
    self.b = np.random.randint(0, 10)

    for i in range(self.iter):
      y_hat = self.predict(X)
      dW = -2 * ((X.T).dot(Y - y_hat)) / row # we will not use sum() as X.dot() will by itself so sum between elements as it is dot product 
      db = -2 * sum(Y - y_hat) / row

      self.W = self.W - self.learning_rate * dW
      self.b = self.b - self.learning_rate * db

      plt.scatter(X,Y)
      plt.plot(X,y_hat , color = 'r')
      plt.title(str(i))
      plt.show()

# mathermatically X.dot(w) and w.dot(X) is same but w.dot(X) would give us some error due to broadcasting ( as x dimesn maybe more than w's)

  def predict(self , X):
    'X-> X_test'
    return X.dot(self.W) + self.b
```

```python
gd = CustomLinearRegressionGD(0.01 , 10)
```
