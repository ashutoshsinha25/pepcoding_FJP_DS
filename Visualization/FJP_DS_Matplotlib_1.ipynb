## Matplotlib for beginners 

* Create a variety of charts, Bar Charts, Line Charts, Stacked Charts, Donut and Pie Charts, Histograms, KDE plots, Violinplots, Boxplots, Scatter Plots, Heatmaps..etc
* Feel comfortable managing various Matplotlib Artists such as Legends, Annotations, Texts, Patches, Lines, Collections, Containers, Axis



**Architecture of matplotlib:**
The Matplotlib architecture is composed of three main layers:


* **Scripting Layer** — The lightest scripting interface among the three layers, designed to make Matplotlib work like MATLAB script.
* **Artist Layer** — Allows full control and fine-tuning of the Matplotlibfigure — the top-level container for all plot elements.
* **Backend Layer** — Handles all the heavy works via communicating to the drawing toolkits in your machine. It is the most complex layer.




**Back-end layer:**
Back-end layer of matplotlib provides the implementations of three abstract interface classes.

1. FigureCanvas:
Provides area onto which figure is drawn.
Example: for drawing in real world, we need paper onto which we can draw. FigureCanvas is similar to paper.
>matplotlib.backend_bases.FigureCanvas

2. Renderer:
It does drawing on FigureCanvas.Instance of Renderer knows how to draw on the FigureCanvas.
Example: just like we need paintbrush , pen or pencil to draw on paper , we use Renderer object to draw on FigureCanvas.
>matplotlib.backend_bases.Renderer

3. Event:
Instance of Event handles user input like keystrokes, mouse clicks.
>matplotlib.backend_bases.Event

**Artist layer:**
* Artist layer comprised of one object known as Artist.
* FigureCanvas from back-end layer is paper, artist object knows how to use Renderer object to draw on canvas.
* Everything we see on matplotlib figure is an instance of Artist. Example: title,lines,labels,images etc.

**Scripting layer:**
* Artist layer is syntactically heavy for the everyday purposes, especially for exploratory work . Matplotlib provides lighter scripting interface to simplify common task .
* Scripting layer composed mainly of pyplot, a lighter interface than Artist layer .
> matplotlib.pyplot

```python
import matplotlib 
matplotlib.__version__
```

```python
# importing lib with alias
import matplotlib.pyplot as plt 
```

```python
#  The plot() function is used to draw points in a figure. 
#  The show() function is used to show the figure.

x = [10 ,20, 30]
y = [20 ,40, 30]

plt.plot(x, y)
plt.show()
```

```python
# If we do not specify the points in the x-axis, they will get the default values 0, 1, 2, 3.
y = [10, 40, 20, 30]
plt.plot(y)
plt.show()
```

```python
# We can call help() to know more about the plot()
help(plt.plot)
```

```python
# we can use the keyword argument markersize or the shorter version, ms to set the size of the markers
# we can use the keyword argument markeredgecolor or the shorter mec to set the color of the edge of the markers
# we can use the keyword argument markerfacecolor or the shorter mfc to set the color inside the edge of the markers
plt.plot(y, marker = 'o', ms = 20, mec = 'red', mfc = 'orange')
plt.show()
```

```python
# We can use the keyword argument color or the shorter c to set the color of the line
# we can use the keyword argument linewidth or the shorter lw to change the width of the line
plt.plot(y, ls = '--', c = 'orange', lw = '3')
plt.show()
```

**Format String**
* We can use also use the shortcut string notation parameter to specify the marker.
* This is also called fmt, and is written with this syntax:
>marker | line | color

```python
plt.plot(y, '^:r')
plt.show()
```

```python
# we can use the xlabel() and ylabel() functions to set a label for the x- and y-axis.
# the title() function is used to set a title for the plot.
# the grid() function is used to add grid lines to the plot.


x = [10, 11, 12, 13, 14]
y = [12, 14, 20, 22, 30]

plt.plot(x, y, marker = 'o')

plt.title("Temperature Data")
plt.xlabel("Time")
plt.ylabel("Temperature in cel")

plt.grid()

plt.show()
```

```python
import numpy as np
y = np.linspace(10, 30, 15)

plt.plot(y)
plt.plot(y**2.5)
plt.plot(y**3)
plt.show()
```

```python
plt.figure()
plt.plot(y)
plt.figure()
plt.plot(y**2.5)
plt.figure()
plt.plot(y**3)
plt.show()
```

```python
# changing the figure size 
plt.figure(figsize=(12, 4))
plt.plot(y**3)
```
