## Slicing of pandas series
  - implecit : index iloc [start , end) ---> end not included
  - explicit : index loc [start,end] ---> end included

```python
a
```

```python
# implicit
a.iloc[0:2]
```

```python
# explicit
a.loc["a":'c']
```

## Handling missing data

```python
df = pd.DataFrame({"Height":{"A":48,"C":69},
                   "Weight":{"B":5.8,"C":6.2},
                   "Age":{"A":10,"B":12,"C":15}
                   })
df
```

```python
df.dropna(axis=1)
```

```python
df['Height'].fillna(df['Height'].mean(),inplace= True)
```

```python
df
```

```python
# df.loc["A"].fillna(30,inplace=True)
df.loc[["A"],:].fillna(30,inplace = True)
```

```python
df
```
